```python
import opendatasets as od

URL = 'https://www.kaggle.com/prathamsharma123/farmers-protest-tweets-dataset-raw-json'
od.download(URL)
```

```python
import os

PATH = 'farmers-protest-tweets-dataset-raw-json'
os.rename(PATH, 'db')
```

```python
import pandas as pd

df = pd.read_json('db/farmers-protest-tweets-2021-03-5.json', lines=True)
df['username'] = df['user'].apply(lambda x: x['username'])
df
```

**1)** Top 10 retweeted tweets

```python
df.sort_values(by=['retweetCount'], ascending=False)[['username', 'retweetCount', 'content']][0:10]
```

**2)** Top 10 users who tweet the most

```python
df.groupby(['username'])['username'].count().sort_values(ascending=False)[0:10]
```

**3)** Top 10 most active days

```python
df['date'] = df['date'].apply(lambda x: x.date())
df.groupby(['date'])['date'].count().sort_values(ascending=False)[0:10]
```

**4)** Top 10 more used hashtags

```python
import re

df['hashtag'] = df['renderedContent'].apply(lambda x: re.findall(r"#(\w+)", x))
df['hashtag']
```

```python
dicc = {}
def count_hashtags(list_):
    for element in list_:
        if element not in dicc:
            dicc[element] = 1
        else:
            dicc[element] += 1

df['hashtag'].apply(count_hashtags)

dictlist = [ [key, value] for key, value in dicc.items() ]
dictlist.sort(key=lambda x: x[1], reverse=True)
dictlist[0:10]
```
