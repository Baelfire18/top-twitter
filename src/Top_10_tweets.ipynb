```python
import os

PATH = 'farmers-protest-tweets-dataset-raw-json'
if os.path.isdir(PATH):
    os.rename(PATH, 'db')
```

```python
import pandas as pd

df = pd.read_json('db/farmers-protest-tweets-2021-03-5.json', lines=True)
df['username'] = df['user'].apply(lambda x: x['username'])
df.rename(columns = {'date':'datetime'}, inplace = True)
df.head()
```

**1)** Top 10 retweeted tweets

```python
def top_10_retweeted_tweets(df):
    return df.sort_values(by=['retweetCount'], ascending=False)[['username', 'retweetCount', 'content']][0:10]

top_10_retweeted_tweets(df)
```

**2)** Top 10 users who tweet the most

```python
def make_serie_into_df(s, column_0):
    return pd.DataFrame([s.index, s.values]).T.rename(columns = {0: column_0, 1: 'number_of_tweets'})
```

```python
def top_10_users_with_most_tweets(df):
    answer_in_serie = df.groupby(['username'])['username'].count().sort_values(ascending=False)
    return make_serie_into_df(answer_in_serie, column_0='username')[0:10]

top_10_users_with_most_tweets(df)
```

**3)** Top 10 most active days

```python
def top_10_active_days(df):
    df['date'] = df['datetime'].apply(lambda x: x.date())
    answer_in_serie = df.groupby(['date'])['date'].count().sort_values(ascending=False)
    return make_serie_into_df(answer_in_serie, column_0='date')[0:10]

top_10_active_days(df)
```

**4)** Top 10 more used hashtags

```python
import re

def top_10_hashtags_more_used(df):
    df['hashtag'] = df['renderedContent'].apply(lambda x: re.findall(r"#(\w+)", x))
    
    hashtag_counter = {}
    def count_hashtags(list_):
        for element in list_:
            if element not in hashtag_counter:
                hashtag_counter[element] = 1
            else:
                hashtag_counter[element] += 1
    df['hashtag'].apply(count_hashtags)

    dictlist = [ [key, value] for key, value in hashtag_counter.items() ]
    dictlist.sort(key=lambda x: x[1], reverse=True)
    
    df_hashtags = pd.DataFrame(dictlist).rename(columns = {0: 'hashtag', 1: 'number_of_tweets'})
    return df_hashtags[0:10]

top_10_hashtags_more_used(df)
```

```python
def validate_input(int_input):
    try:
        if int(int_input) not in list(range(1, 5)):
            int_input = input("Please enter a valid option.\n> ")
            return validate_input(int_input)
        return int(int_input)
    except ValueError:
        int_input = input("Please enter a valid option.\n> ")
        return validate_input(int_input)


def main():
    if True:
        options_str = """Available functions:
    [1] Top 10 most retweeted tweets
    [2] Top 10 users that tweeted the most
    [3] Top 10 most active days
    [4] Top 10 most used hashtags"""
    
        functions_dict = [top_10_retweeted_tweets,
                          top_10_users_with_most_tweets,
                          top_10_active_days,
                          top_10_hashtags_more_used]

        selected_function = validate_input(input(f"{options_str}\n> "))
        result = functions_dict[selected_function - 1](df)
        print("Execution result:\n")
        return result
```

```python
if __name__ == "__main__":
    main()
```
